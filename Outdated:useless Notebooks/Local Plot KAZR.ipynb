```python
import numpy as np
from scipy.io import netcdf
f = netcdf.netcdf_file('sgpkazrgeC1.a1.20140527.000002.cdf', 'r')
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
%matplotlib inline
```

```python
# Read netCDF variables
rxn = f.variables['rx_noise'].data
refc = f.variables['reflectivity_copol'].data
refx = f.variables['reflectivity_xpol'].data
mdvc = f.variables['mean_doppler_velocity_xpol'].data
mdvx = f.variables['mean_doppler_velocity_copol'].data
snx = f.variables['sky_noise_xpol'].data
snc = f.variables['sky_noise_copol'].data
alt = f.variables['alt'].data
hnc = f.variables['hot_noise_copol'].data
cnc = f.variables['cold_noise_copol'].data
hnx = f.variables['hot_noise_xpol'].data
cnx = f.variables['cold_noise_xpol'].data 
lon = f.variables['lon'].data
lat = f.variables['lat'].data
stnrx = f.variables['signal_to_noise_ratio_xpol'].data
stnrc = f.variables['signal_to_noise_ratio_copol'].data
swc = f.variables['spectral_width_copol'].data
swx = f.variables['spectral_width_xpol'].data
ccc = f.variables['cal_constant_copol'].data
ccx = f.variables['cal_constant_xpol'].data
rng = f.variables['range'].data
btime = f.variables['base_time'].data
time = f.variables['time'].data
timeo = f.variables['time_offset'].data
```

```python
fig = plt.figure(figsize = [15,8])
plt.title( 'Copol Reflectivity 5/27/2014' )
plt.xlabel('UTC Time (Hours)',fontsize=14)
plt.ylabel('Altitude (Meters)',fontsize=14)


#CONVERSION?

p_s = 101325 #Pa
h = f.variables['range'][:] #m
T = 288.15 #K
g = 9.80665 #m/s2
R = 8.31447 #J/molK
M = 0.0289644 #kg/mol
L = 0.0065 #K/m
                          
p = [(p_s*((1-((L*h[i])/T))**((g*M)/(R*L)))/100) for i in range(len(h))]
#pc = plt.pcolormesh(np.array(day), np.array(p), f.variables['reflectivity_copol'][:].transpose(), cmap='jet')

#ax.set_ylim([max(p),0])

from mpl_toolkits.axes_grid1 import make_axes_locatable

t = f.variables['time'][:]
day = [t[i]/3600 for i in range(len(t))]

#axis 1
ax1 = plt.gca()
ax1.set_ylim([0,max(rng)])
ax1.set_xlim([0, 24])
majorLocator = MultipleLocator(1)
#minorLocator = MultipleLocator(0.5)
ax1.xaxis.set_major_locator(majorLocator)
#ax.xaxis.set_minor_locator(minorLocator)
majorFormatter = FormatStrFormatter('%d')
ax1.xaxis.set_major_formatter(majorFormatter)

pc1 = plt.pcolormesh(np.array(day), rng, f.variables['reflectivity_copol'][:].transpose(), cmap='jet')

#axis2
ax2 = ax1.twinx()
ax2.set_ylim([max(p),0])
ax2.set_xlim([0, 24])
majorLocator = MultipleLocator(1)
#minorLocator = MultipleLocator(0.5)
ax2.xaxis.set_major_locator(majorLocator)
#ax.xaxis.set_minor_locator(minorLocator)
majorFormatter = FormatStrFormatter('%d')
ax2.xaxis.set_major_formatter(majorFormatter)

cb = plt.colorbar(mappable=pc1)
cb.set_label(r'Reflectivity factor, $Z_e$ (dBZ)')
#plt.tight_layout()
#plt.colorbar()
```
