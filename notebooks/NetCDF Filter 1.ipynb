```python
import numpy as np
from scipy.io import netcdf
import pyhrrr
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

# Gather files and capture date
radar = 'sgpkazrgeC1.a1.20140527.000002.cdf'
sounding = 'sgpsondewnpnC1.b1.20140527.052800.custom.cdf'
year = int(radar[15:19])
month = int(radar[19:21])
day = int(radar[21:23])
f = netcdf.netcdf_file(radar, 'r')
s = netcdf.netcdf_file(sounding, 'r')

%matplotlib inline
```

```python
#Define Radar and Sounding Heights and Pressure
rng = f.variables['range'][:]

psnd = s.variables['pres'][:]
hsnd = s.variables['alt'][:]
hrdr = (rng)+320 # Correction for height above mean sea level
prdr = []

#Interpolation of radar pressure from the balloon sounding data
prdr = np.interp(hrdr,hsnd,psnd)
```

```python
#Use 'get_sun' function to find decimal UTC time for sunrise and sunset
date = datetime.datetime(year,month,day)
x = pyhrrr.get_sun(date=date)
trise = x[0][0]
tset = x[0][1]
```

```python
fig = plt.figure(figsize = [15,8])
plt.title( 'SGP KAZR Copol Reflectivity 5/27/2014' )
plt.xlabel('UTC Time (Hour)',fontsize=14)
plt.ylabel('Pressure (hPa)',fontsize=14)

#From seconds to hours
t = f.variables['time'][:]
day = [t[i]/3600 for i in range(len(t))]

# set axes scales and limits
ax1 = plt.gca()
ax1.set_yscale('log')
ax1.set_ylim([max(prdr),min(prdr)])
ax1.set_xlim([0, 24])

# x axis
xmajorLocator = MultipleLocator(1)
ax1.xaxis.set_major_locator(xmajorLocator)
xmajorFormatter = FormatStrFormatter('%d')
ax1.xaxis.set_major_formatter(xmajorFormatter)

# y axis
ymajorLocator = MultipleLocator(100)
ax1.yaxis.set_major_locator(ymajorLocator)
ymajorFormatter = FormatStrFormatter('%d')
ax1.yaxis.set_major_formatter(ymajorFormatter)

# plot radar reflectivity
pc1 = plt.pcolormesh(np.array(day), prdr, f.variables['reflectivity_copol'][:].transpose(), cmap='jet')

# draw vertical lines for sunrise and sunset
ax1.axvline(trise, linestyle='--', color='k')
ax1.axvline(tset, linestyle='--', color='k')
plt.gca()
ax1.text(trise,100,'Sunrise')
ax1.text(tset,100,'Sunset')

# Plot Colorbar
cb = plt.colorbar(ax=ax1,mappable=pc1,orientation='vertical')
cb.set_label(r'Reflectivity factor, $Z_e$ (dBZ)')
```

```python
import scipy.stats as st

fig = plt.figure(figsize = [12.3,8])

# Calculate SNR
copol = (f.variables['reflectivity_copol'][:])
snr = st.signaltonoise(copol,axis = 0)

#Label Figure
plt.title('Signal to Noise Ratio')
plt.ylabel('Height (m)')
plt.xlabel('SNR')

# set axes scales and limits
ax1 = plt.gca()
ax1.set_ylim([0,max(rng)])

# y axis
#ymajorLocator = MultipleLocator(25000)
#ax1.yaxis.set_major_locator(ymajorLocator)
#ymajorFormatter = FormatStrFormatter('%d')
#ax1.yaxis.set_major_formatter(ymajorFormatter)

# Plot Line
plt.plot(snr,rng, color = 'm', linewidth = 2)
plt.grid()
```

```python
fig = plt.figure(figsize = [15,8])
plt.title( 'SGP KAZR LDR Reflectivity 5/27/2014' )
plt.xlabel('UTC Time (Hour)',fontsize=14)
plt.ylabel('Pressure (hPa)',fontsize=14)

# Convert seconds to hours for x axis
t = f.variables['time'][:]
day = [t[i]/3600 for i in range(len(t))]

# set axes scales and limits
ax1 = plt.gca()
ax1.set_yscale('log')
ax1.set_ylim([max(prdr),min(prdr)])
ax1.set_xlim([0, 24])

# x axis
xmajorLocator = MultipleLocator(1)
ax1.xaxis.set_major_locator(xmajorLocator)
xmajorFormatter = FormatStrFormatter('%d')
ax1.xaxis.set_major_formatter(xmajorFormatter)

# y axis
ymajorLocator = MultipleLocator(100)
ax1.yaxis.set_major_locator(ymajorLocator)
ymajorFormatter = FormatStrFormatter('%d')
ax1.yaxis.set_major_formatter(ymajorFormatter)

# Calculate Linear Depolarization Ratio
copol = (f.variables['reflectivity_copol'][:])
xpol = (f.variables['reflectivity_xpol'][:])
ldr = (xpol-copol)

# Plot LDR
pc2 = plt.pcolormesh(np.array(day), prdr, ldr.transpose(), cmap='jet')

# Plot Colorbar
cb = plt.colorbar(ax=ax1,mappable=pc2,orientation='vertical')
cb.set_label(r'Reflectivity factor, $Z_e$ (dBZ)')
```

```python
import numpy.ma as ma

fig = plt.figure(figsize = [15,8])
plt.title( 'Insect Filtered' )
plt.xlabel('UTC Time (Hour)',fontsize=14)
plt.ylabel('Pressure (hPa)',fontsize=14)

#From seconds to hours
t = f.variables['time'][:]
day = [t[i]/3600 for i in range(len(t))]

# set axes scales and limits
ax1 = plt.gca()
ax1.set_yscale('log')
ax1.set_ylim([max(prdr),min(prdr)])
ax1.set_xlim([0, 24])

# x axis
xmajorLocator = MultipleLocator(1)
ax1.xaxis.set_major_locator(xmajorLocator)
xmajorFormatter = FormatStrFormatter('%d')
ax1.xaxis.set_major_formatter(xmajorFormatter)

# y axis
ymajorLocator = MultipleLocator(100)
ax1.yaxis.set_major_locator(ymajorLocator)
ymajorFormatter = FormatStrFormatter('%d')
ax1.yaxis.set_major_formatter(ymajorFormatter)

# Calculate Linear Depolarization Ratio
copol = (f.variables['reflectivity_copol'][:])
xpol = (f.variables['reflectivity_xpol'][:])
ldr = (xpol-copol)

# Plot Copolar Reflectivity with LDR mask
new = ma.masked_where(np.logical_and(ldr >= -10, prdr >= 800),copol)
pc2 = plt.pcolormesh(np.array(day), prdr, new.transpose(), cmap='jet')

# Plot Colorbar
cb = plt.colorbar(ax=ax1,mappable=pc2,orientation='vertical')
cb.set_label(r'Reflectivity factor, $Z_e$ (dBZ)')
```
