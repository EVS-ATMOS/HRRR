```python
import numpy as np
import os
import json
import datetime
import matplotlib.dates
import pyhrrr
```

```python
date = datetime.datetime(2014,7,1,0,0)
filenum = 24
hour = 1
directory = '/data/san_store/HRRR'
enddirectory = '/home/mjohnson/python/hrrr_tests'
loc=[36.605, -97.485]
write_modelhours = False
```

```python
wkdir = os.getcwd()
    
#if ((type(hour) == list)):
 #   print 'error, can only write one model hour at a time if write_modelhours = False'
    
    
newfilename = pyhrrr.produce_hrrr_txt_string(date=date,hour=hour,filenum=filenum,loc=loc,indexes =None,modelhours=write_modelhours)

if newfilename in os.listdir(enddirectory):
    print 'error file already exists'

    
#date = date-datetime.timedelta(hours=hour)


datestrings = [date+datetime.timedelta(hours=i) for i in range(24)]
hourslists = [[hour] for i in range(24)]
        
filelists = pyhrrr.produce_hrrr_grib2strings(datestrings,hourslists)

print filelists

if filelists == []:
    print 'warning2'
        
data = []

    
dates = []
k = -1
    
for i in range(len(filelists)):
    if filelists[i] in os.listdir(directory):
        x = pyhrrr.read_hrrr_spec(filename = filelists[i], directory = directory,no_txt = True,coords=None)
        print filelists[i]
        if x != None and i>k:
            k = len(filelists)+1
            parameterlist = x[1]
            loc = x[2]
            indexes = x[3]
            units = x[4]
        if x == None:
            continue
        x[0] = np.array(x[0])
        x[0] = x[0].tolist()
        data.append(x[0])
        x = None
        dates.append(matplotlib.dates.date2num(datestrings[i]))
            
if not ('parameterlist' in vars().keys()):
    print 'parameterlist not in vars'
    
```
