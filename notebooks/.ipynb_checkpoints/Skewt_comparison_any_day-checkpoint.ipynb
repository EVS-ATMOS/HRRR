```python
import pyhrrr
import numpy as np
from scipy.io import netcdf
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import datetime

#Read in balloon sounding
sounding = '/data/san_store/soundings/sgp/mckercherg1/159821/sgpsondewnpnC1.b1.20140527.172800.cdf'
s = netcdf.netcdf_file(sounding, 'r')
```

```python
%matplotlib inline

psnd = s.variables['pres'][:]
hsnd = s.variables['alt'][:]
tsnd = s.variables['tdry'][:]
tdsnd = s.variables['dp'][:]
u = s.variables['u_wind'].data
v = s.variables['v_wind'].data

pyhrrr.plot_skewt_wind(psnd,hsnd,tsnd,tdsnd,u,v)
```

```python
#HRRR data
p_hrrr = pyhrrr.HRRR_PS
[data,dates,parameterlist,loc,indexes,units] = pyhrrr.read_hrrr_txt(date = datetime.datetime(2014,5,27),hour=1,filenum=24,
                            directory = '/data/san_store/hrrr_txt',loc = [36.605,-97.485],
                            indexes=None,read_modelhours=False)
# Find model hour closest to entered time
hour = 1
c_date = datetime.datetime(2014,5,27,17,28,tzinfo=dates[0].tzinfo)-datetime.timedelta(hours=hour)
date_ch = [i-c_date for i in dates]
date_shifts = [i.total_seconds() for i in date_ch]
date_shifts = np.array(date_shifts)**2
ind = date_shifts.tolist().index(min(date_shifts))
# Gather parameters
ti = pyhrrr.HRRR_VARS.index('Temperature')
tdi = pyhrrr.HRRR_VARS.index('Dew point temperature')
ui = pyhrrr.HRRR_VARS.index('U component of wind')
vi = pyhrrr.HRRR_VARS.index('V component of wind')
data=np.array(data)
T_hrrr = (data[:,ti,:])-273.15
T_hrrr = T_hrrr[ind,:]
Td_hrrr = (data[:,tdi,:])-273.15
Td_hrrr = Td_hrrr[ind,:]
U_hrrr = data[:,ui,:]
U_hrrr = U_hrrr[ind,:]
V_hrrr = data[:,vi,:]
V_hrrr = V_hrrr[ind,:]
#Interpolation of HRRR height using sounding
h_hrrr = []
h_hrrr = np.interp(p_hrrr[::-1],psnd[::-1],hsnd[::-1])
h_hrrr = h_hrrr[::-1]
# Plot model skewt
pyhrrr.plot_skewt_wind(p_hrrr,h_hrrr,T_hrrr,Td_hrrr,U_hrrr,V_hrrr)
```
