```python
radar_filename = 'sgpkazrgeC1.a1.20140419.000001.cdf'
sounding_filename = 'sgpsondewnpnC1.b1.20140419.052400.cdf'
radar_directory = '/Users/mattjohnson/'
sounding_directory = '/Users/mattjohnson/'
```

```python
import pyhrrr
import os
import numpy as np
%matplotlib inline
```

```python
produce_file = False
psinds = None
tsinds = None
```

```python
wkdir = os.getcwd()

x = pyhrrr.get_netcdf_variables(filename = radar_filename, directory = radar_directory,variablelist=
                                ['reflectivity_copol','range','time'])

copol = x[0][0][0]    

ran = x[0][0][1]

times = x[0][0][2]
             
[[sdata,sdim,sunits],sdate,f] = pyhrrr.get_netcdf_variables(filename=sounding_filename,directory=sounding_directory,
                                                            variablelist=['pres','alt'])
    

pres = np.interp(ran,sdata[1],sdata[0])
    
if tsinds == None and psinds == None:
    [psinds,tsinds] = pyhrrr.calc_radar2hrrr_inds(times,np.array(pres))
    
    
copol = np.array(copol)
copol = 10**(copol/10)
    
z = []
y = []
    

for i in range(len(tsinds)-1):
    for j in range(len(psinds)-1):
        if psinds[j] != psinds[j+1] and tsinds[i] != tsinds[i+1]:
            temp = float(np.nanmean(np.nanmean(copol[tsinds[i]:tsinds[i+1],psinds[j]:psinds[j+1]],axis=1),axis=0))
            if temp == None:
                temp = np.nan
            y.append(temp)
    z.append(y)
    y = []
        
if z == None:
    print 'z is None out of averaging'

        
z = np.array(z)
z = 10*np.log10(z)
indexes = np.where(z==np.nan)
indexes = np.array(indexes)
z = z.tolist()
    
for i in range(indexes.shape[1]):
    z[indexes[0][i]][indexes[1][i]] = None
    
hrrr_heights = np.interp(pyhrrr.HRRR_PS[::-1],sdata[0][::-1],sdata[1][::-1])
hrrr_heights = hrrr_heights[::-1]
if produce_file:
    os.chdir(output_directory)
    import json
    import datetime
    date = datetime.datetime(int(radar_filename[15:19]),int(radar_filename[19:21]),int(radar_filename[21:23]))
    filestring = produce_radar_txt_string(date)
    g = open(filestring,'w')
    u = [z.tolist(),hrrr_heights.tolist(),tsinds,psinds]
    json.dump(u,g)
    g.close()
    os.chdir(wkdir)
```
