```python
date = datetime.datetime(2014,7,6) #change for date
```

```python
radar_directory = '/data/san_store/radar_txt'
hrrr_directory = '/data/san_store/hrrr_txt'
radar_string = pyhrrr.produce_radar_txt_string(date)
hrrr_string = pyhrrr.produce_hrrr_txt_string(date,filenum = 24,hour = 1,loc=[36.605, -97.485], indexes = None,
                                             modelhours = False)
wkdir = os.getcwd()
os.chdir(hrrr_directory)
hrrrf = open(hrrr_string,'r')
hrrr = json.load(hrrrf)
os.chdir(radar_directory)
radarf = open(radar_string,'r')
radar = json.load(radarf)

hrrr[0] = np.array(hrrr[0])
heights = radar[2]
snr = np.array(radar[1])
#ind = hrrr[2].index() #change for parameter
hrrr_data = np.array(hrrr[0][:,-5,:])+np.array(hrrr[0][:,-4,:])+np.array(hrrr[0][:,-3,:])+np.array(hrrr[0][:,-2,:])+np.array(hrrr[0][:,-1,:])
press_index = pyhrrr.HRRR_PS.tolist().index(800)
#hrrr_c = hrrr_data[:,ind]
radar_data = np.array(radar[0])


radar_m = pyhrrr.filter_mask(radar_data,snr,-14)
radar_i = pyhrrr.filter_mask(radar_data,snr,-14)

ceil_pres = np.interp(radar[2],radar[3],np.array(pyhrrr.HRRR_PS))

#import bisect
#ceil_inds = []
#for i in ceil_pres:
    #ceil_inds.append(bisect.bisect_left(pyhrrr.HRRR_PS.tolist(),i))

for k in range(radar_data.shape[0]):
    if radar[2][k] == 2000.:
        radar_m[k,:] = np.ma.masked_where(radar_m[k,:]!=-5000,radar_m[k,:])
    #else:
        #pass
        #c_ind = ceil_inds[k]
        #temp = radar_m[k,0:c_ind+1]
        #temp2 = radar_m[k,c_ind+1::]
        #radar_m[k,0:c_ind+1,] = np.ma.masked_where(temp<-5000,temp)
        #radar_m[k,c_ind+1::] = np.ma.array(temp2)

    
radar_m = np.ma.masked_where(radar_m==-np.inf,radar_m)


radar_i = np.ma.masked_where(radar_i==-np.inf,radar_i)


hrrr_data_m = pyhrrr.filter_mask(hrrr_data,hrrr_data,10**-5)
rh_m = pyhrrr.filter_mask(hrrr[0][:,2,:],hrrr[0][:,2,:],80)
```
