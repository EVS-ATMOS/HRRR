```python
import numpy as np
from scipy.io import netcdf
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

# Gather files and capture date
radar = 'sgpkazrgeC1.a1.20140527.000002.cdf'
sounding = 'sgpsondewnpnC1.b1.20140527.052800.custom.cdf'
ceilometer = 'sgpvceil25kC1.b1.20140527.000007.cdf'
f = netcdf.netcdf_file(radar, 'r')
s = netcdf.netcdf_file(sounding, 'r')
c = netcdf.netcdf_file(ceilometer, 'r')

%matplotlib inline
```

```python
fcbh = c.variables['first_cbh'].data
scbh = c.variables['second_cbh'].data
tcbh = c.variables['third_cbh'].data
```

```python
#From seconds to hours
t = c.variables['time'][:]
day = [t[i]/3600 for i in range(len(t))]
```

```python
rng = f.variables['range'].data
```

```python
refc = f.variables['reflectivity_copol'].data
refx = f.variables['reflectivity_xpol'].data
```

```python
fig = plt.figure(figsize = [12,8])
plt.scatter(day,fcbh, marker='.')
plt.scatter(day,scbh, marker='.', color='r')
plt.scatter(day,tcbh, marker='.', color='g')
plt.legend(('First Cloud Base Height','Second Cloud Base Height','Third Cloud Base Height'))
plt.title( 'SGP Ceilometer Cloud Base Height 5/27/2014' )
plt.xlabel('UTC Time (Hours)',fontsize=14)
plt.ylabel('Height (m)',fontsize=14)

ax = plt.gca()
ax.set_ylim(0,max(rng))
ax.set_xlim([0, 24])

# x axis
majorLocator = MultipleLocator(1)
ax.xaxis.set_major_locator(majorLocator)
majorFormatter = FormatStrFormatter('%d')
ax.xaxis.set_major_formatter(majorFormatter)

# y axis
ymajorLocator = MultipleLocator(2500)
ax.yaxis.set_major_locator(ymajorLocator)
ymajorFormatter = FormatStrFormatter('%d')
ax.yaxis.set_major_formatter(ymajorFormatter)

plt.grid()
plt.show()
```

```python
import numpy.ma as ma

fig = plt.figure(figsize = [15,8])
plt.title( 'SGP KAZR Copol Reflectivity / Ceilometer Cloud Base 5/27/2014' )
plt.xlabel('UTC Time (Hour)',fontsize=14)
plt.ylabel('Height (m)',fontsize=14)

# set axes scales and limits
ax1 = plt.gca()
ax1.set_ylim([min(rng),max(rng)])
ax1.set_xlim([0, 24])

# x axis
xmajorLocator = MultipleLocator(1)
ax1.xaxis.set_major_locator(xmajorLocator)
xmajorFormatter = FormatStrFormatter('%d')
ax1.xaxis.set_major_formatter(xmajorFormatter)

# y axis
ymajorLocator = MultipleLocator(2500)
ax1.yaxis.set_major_locator(ymajorLocator)
ymajorFormatter = FormatStrFormatter('%d')
ax1.yaxis.set_major_formatter(ymajorFormatter)

# Filtering
stnrc = f.variables['signal_to_noise_ratio_copol'].data
refc = ma.masked_where((stnrc <= -14),refc)

# Calculate Linear Depolarization Ratio
ldr = (refx-refc)

# plot radar reflectivity
t = f.variables['time'][:]
day = [t[i]/3600 for i in range(len(t))]
pc0 = plt.pcolormesh(np.array(day), rng, refc.T, cmap='jet')

# Plot Colorbar
cb = plt.colorbar(ax=ax1,mappable=pc0,orientation='vertical')
cb.set_label(r'Reflectivity factor, $Z_e$ (dBZ)')

ct = c.variables['time'][:]
cday = [ct[i]/3600 for i in range(len(ct))]
plt.scatter(cday,fcbh,s=2, marker='.')
```

```python
import numpy.ma as ma

fig = plt.figure(figsize = [15,8])
plt.title( 'SGP KAZR LDR Reflectivity 5/27/2014' )
plt.xlabel('UTC Time (Hour)',fontsize=14)
plt.ylabel('Height (m)',fontsize=14)

# set axes scales and limits
ax1 = plt.gca()
ax1.set_ylim([min(rng),max(rng)])
ax1.set_xlim([0, 24])

# x axis
xmajorLocator = MultipleLocator(1)
ax1.xaxis.set_major_locator(xmajorLocator)
xmajorFormatter = FormatStrFormatter('%d')
ax1.xaxis.set_major_formatter(xmajorFormatter)

# y axis
ymajorLocator = MultipleLocator(2500)
ax1.yaxis.set_major_locator(ymajorLocator)
ymajorFormatter = FormatStrFormatter('%d')
ax1.yaxis.set_major_formatter(ymajorFormatter)

refc = ma.masked_where(np.logical_and(ldr >= -20, rng <= 2000),refc)
pc1 = plt.pcolormesh(np.array(day), rng, refc.T, cmap='jet')

# Plot Colorbar
cb = plt.colorbar(ax=ax1,mappable=pc1,orientation='vertical')
cb.set_label(r'Reflectivity factor, $Z_e$ (dBZ)')
```
