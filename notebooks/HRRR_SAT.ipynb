```python
import pyart, pygrib
import sys, os, urllib, urllib2
from matplotlib import pyplot as plt
from matplotlib import colors
import numpy as np
from mpl_toolkits.basemap import Basemap, addcyclic, pyproj
import netCDF4

%matplotlib inline
```

```python

def fetch_HRRR(forecast_time = 'f000', dimensionality = '3d', outdir = '/data/HRRR/'):
    HRRR_url = 'http://hrrr.agron.iastate.edu/data/hrrr/'
    all_files = urllib.urlopen(HRRR_url).read()
    all_times = []
    for item in all_files.split("\n"):
        if "a href" in item:
            idx = item.find('a href="')
            all_times.append(item[idx + 8 : idx + 20])
    last_dir = HRRR_url + all_times[-1]
    print last_dir
    filename = 'hrrr.' + dimensionality + '.' + all_times[-1] + forecast_time + '.grib2'
    exising_hrrr_files = os.listdir(outdir)
    outfile = outdir + filename
    if filename in exising_hrrr_files:
        print('File exists, saving bandwidth')
    else:
        print('fetching')
        try:    
            req = urllib2.urlopen(last_dir + '/' + filename)
            outobj=open(outfile, 'wb')
            outobj.write(req.read())
            outobj.close()
        except:
            outfile = 'null'
            print('File not found, skipping')
    return outfile


def grb_to_grid(grb_obj):
    """Takes a single grb object containing multiple
    levels. Assumes same time, pressure levels. Compiles to a cube"""
    n_levels = len(grb_obj)
    levels = np.array([grb_element['level'] for grb_element in grb_obj])
    indexes = np.argsort(levels)[::-1] # highest pressure first
    cube = np.zeros([n_levels, grb_obj[0].values.shape[0], grb_obj[1].values.shape[1]])
    for i in range(n_levels):
        cube[i,:,:] = grb_obj[indexes[i]].values
    cube_dict = {'data' : cube, 'units' : grb_obj[0]['units'],
                 'levels' : levels[indexes]}
    return cube_dict


def max_IWMR_plot(grb_cube, lats, lons):
    max_ice = grb_cube['data'].max(axis=0)
    f = plt.figure(figsize=[12,10])
    m = Basemap(llcrnrlon = -110,llcrnrlat = 20, urcrnrlon = -70,
               urcrnrlat = 52 , projection = 'mill', area_thresh =1000 ,
               resolution='l')
    x, y = m(lons, lats)
    
    my_mesh = m.pcolormesh(x, y, max_ice*1000.0, vmax =  .01, vmin = 1e-6)
    my_coast = m.drawcoastlines(linewidth=1.25)
    my_states = m.drawstates()
    m.drawparallels(np.linspace(10,50, 9) ,labels=[1,1,0,0])
    m.drawmeridians(np.linspace(-110, -80,7),labels=[0,0,0,1]) 
    plt.colorbar(label='g/m^3')   
```

```python
dir_string = 'http://motherlode.ucar.edu:8080/thredds/dodsC/satellite/IR/EAST-CONUS_4km/current/'
dirlisting = urllib2.urlopen(dir_string)
all_lines = dirlisting.read().splitlines()
has_gini = []
for line in all_lines:
    if '.gini' in line:
        #print line
        happy_place = line.find('EAST-CONUS_4km_IR_')
        end_place = line.find('.gini')
        my_upper_name = line[happy_place:end_place]+'.gini'
        url = dir_string + my_upper_name
        has_gini.append(url)
```

```python
print has_gini[0]
data=netCDF4.Dataset(has_gini[0])
```

```python
hrr_file = fetch_HRRR(forecast_time = 'f001')
```

```python
xg, yg = np.meshgrid(data.variables['x'][:]*1000.0, data.variables['y'][:]*1000.0)
vals=np.uint8(data.variables['IR'][:])*-1.0
#swap_me = np.where(vals < 0)
#vals[swap_me] = vals[swap_me] + 128.0*2.0

print vals.shape

fig=plt.figure(figsize=(10,9))
ax = fig.add_axes([0.01,0.05,0.9,0.80]) 
pnyc = pyproj.Proj(proj = 'lcc', 
                   lat_1 = data.variables['LambertConformal'].latitude_of_projection_origin,
                   lat_2 = data.variables['LambertConformal'].latitude_of_projection_origin,
                   lat_0 = data.variables['LambertConformal'].latitude_of_projection_origin,
                   lon_0 = data.variables['LambertConformal'].longitude_of_central_meridian )

lon, lat = pnyc(xg, yg, inverse = True)

min_lon = lon.min(); max_lon = lon.max()
min_lat = lat.min(); max_lat = lat.max()

m = Basemap(llcrnrlon = -110,llcrnrlat = 20, urcrnrlon = -70,
               urcrnrlat = 52 , projection = 'mill', area_thresh =1000 ,
               resolution='h')

x, y = m(lon, lat)
# create figure.
m.drawparallels(np.linspace(10,50, 9) ,labels=[1,1,0,0])
m.drawmeridians(np.linspace(-110, -80,7),labels=[0,0,0,1]) 
m.pcolormesh(x, y , vals[0,:], cmap=plt.get_cmap('binary'))

m.drawcoastlines(linewidth=1.25)
m.drawstates()
plt.colorbar()
```

```python
myfile = pygrib.open(hrr_file)
grb_ice = myfile.select(name='Cloud Ice')
grb_cube_ice=grb_to_grid(grb_ice)
lats, lons = grb_ice[0].latlons()
max_IWMR_plot(grb_cube_ice, lats, lons)
plt.title(r'Max $q_{IWC}$ in col.')
myfile.close()
```
